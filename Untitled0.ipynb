```python
flower=[]
for i in os.listdir(data_dir):
  flower.append(i)
```

```python
#Putting all the images and there corresponding labels into two variable respectively.
import cv2 as cv
import numpy as np
from random import shuffle
#Total_data = []
Total_data1 =[]####
labels = []
def create_train():

  for a in flower:
    path = os.path.join(data_dir,a)
    label = flower.index(a)
    
    for img in os.listdir(path):
      img_path = os.path.join(path,img)
      img_array = cv.imread(img_path)
      #gray = cv.cvtColor(img_array, cv.COLOR_BGR2GRAY)
      gray = cv.resize(img_array, (224,224))      #Resizing Of image to (224X224)
      Total_data1.append(gray)####
      #Total_data.append([gray,label])
      labels.append(label)

create_train()
#shuffle(Total_data)
```

```python
#Printing the Total Image
len(Total_data1),len(labels)
```

```python
#Splitting total data into train and test data 
from sklearn.model_selection import train_test_split
XX_train, XX_test, yy_train, yy_test = train_test_split(Total_data1, labels, test_size=0.3, random_state=42)
from tensorflow.keras.utils import to_categorical
#train = Total_data[:-500]
#test = Total_data[-500:]
#X_train = np.array([i[0] for i in train]).reshape(-1 , 64, 64 ,1)
#y_train = np.array([i[1] for i in train])
#X_test = np.array([i[0] for i in test]).reshape(-1 , 64, 64 ,1)
#y_test = np.array([i[1] for i in test])
#y_train=to_categorical(y_train)
#y_test=to_categorical(y_test)
yy_train=to_categorical(yy_train)    #One Hot encoding
yy_test=to_categorical(yy_test)
XX_train = np.array(XX_train)
XX_test = np.array(XX_test)
XX_train = XX_train.reshape(-1 , 224, 224 ,3)
X_test = XX_test.reshape(-1 , 224, 224,3)
```

BUILDING THE MODEL

```python
import tensorflow as tf
import tflearn
from tflearn.layers.conv import conv_2d,max_pool_2d
from tflearn.layers.core import input_data, dropout, fully_connected
from tflearn.layers.estimator import regression
```

```python
tf.compat.v1.reset_default_graph()
convnet = input_data(shape= [None , 150 , 150 , 3] ,name='input')
convnet = conv_2d(convnet, nb_filter = 64, filter_size = (5,5), strides=2, activation = 'leaky_relu',weights_init='truncated_normal')
convnet = max_pool_2d(convnet, 5,strides=(2,2))
convnet = conv_2d(convnet, nb_filter = 64, filter_size = (3,3), strides=2, activation = 'leaky_relu',weights_init='truncated_normal')
convnet = max_pool_2d(convnet, 5,strides=(2,2))
convnet = conv_2d(convnet, nb_filter = 96, filter_size = (3,3), strides=2, activation = 'leaky_relu',weights_init='truncated_normal')
convnet = max_pool_2d(convnet, 5,strides=(2,2))
convnet = conv_2d(convnet, nb_filter = 128, filter_size = (3,3), strides=2, activation = 'leaky_relu',weights_init='truncated_normal')
convnet = max_pool_2d(convnet, 5,strides=(2,2))
convnet = conv_2d(convnet, nb_filter = 128, filter_size = (3,3), strides=2, activation = 'leaky_relu',weights_init='truncated_normal')

convnet = max_pool_2d(convnet, 5,strides=(2,2))
convnet = fully_connected(convnet , 512, activation = 'leaky_relu')
convnet = dropout(convnet, 0.8)     #For Solving overfitting problem
convnet = fully_connected(convnet , 5, activation = 'softmax')
convnet = regression(convnet, optimizer = 'adam', learning_rate = 0.001, loss = 'categorical_crossentropy',batch_size=200)
model = tflearn.DNN(convnet, tensorboard_dir='log', tensorboard_verbose=0)
model.fit(XX_train , yy_train , n_epoch = 40,
          validation_set=(X_test, yy_test),
          snapshot_step = 500, show_metric = True, run_id ='CNN')
```

```python
import matplotlib.pyplot as plt   # pip install matplotlib
shuffle(Total_data1)

fig = plt.figure(figsize=(20,20))
num1=0
for data in Total_data1[:20]:
    img_data = data
    y = fig.add_subplot(5,5, num1+1)
    num1+=1
    image = img_data
    data = img_data.reshape(150,150,3)
    model_out = model.predict([data])[0]
    
    if np.argmax(model_out) == 0:
        my_label = flower[0]
    elif np.argmax(model_out) == 1:
        my_label = flower[1]
    elif np.argmax(model_out) == 2:
        my_label = flower[2]
    elif np.argmax(model_out) == 3:
        my_label = flower[3]
    else:
        my_label = flower[4]
    
    y.imshow(cv.cvtColor(image,cv.COLOR_BGR2RGB))
    plt.title(my_label)
    
    y.axes.get_xaxis().set_visible(False)
    y.axes.get_yaxis().set_visible(False)
plt.show()
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout, Flatten, Activation
from keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import Adam

def build_model():
  model1 = Sequential()
  model1.add(Conv2D(filters = 64, kernel_size = (5,5),strides=2,activation ='leaky_relu',padding='same',kernel_initializer='truncated_normal', input_shape = (224,224,3)))
  model1.add(MaxPooling2D(pool_size=5,strides=(2,2),padding='same'))

  model1.add(Conv2D(filters = 64, kernel_size = (3,3),strides=2,activation ='leaky_relu',padding='same',kernel_initializer='truncated_normal',kernel_regularizer=keras.regularizers.l2(0.001)))
  model1.add(MaxPooling2D(pool_size=5,strides=(2,2),padding='same'))

  model1.add(Conv2D(filters = 96, kernel_size = (3,3),strides=2,activation ='leaky_relu',padding='same',kernel_initializer='truncated_normal',kernel_regularizer=keras.regularizers.l2(0.001)))
  model1.add(MaxPooling2D(pool_size=5, strides=(2,2),padding='same'))

  model1.add(Conv2D(filters =128, kernel_size = (3,3),strides=2,activation ='leaky_relu',padding='same',kernel_initializer='truncated_normal',kernel_regularizer=keras.regularizers.l2(0.001)))
  model1.add(MaxPooling2D(pool_size=5, strides=(2,2),padding='same'))
  model1.add(Dropout(0.6))
  
  #model1.add(Conv2D(filters = 128, kernel_size = (3,3),strides=2,activation ='relu',kernel_initializer='truncated_normal'))
  #model1.add(MaxPooling2D(pool_size=5, strides=(2,2)))

  model1.add(Flatten())
  model1.add(Dense(512))
  model1.add(Activation('leaky_relu'))
  model1.add(Dropout(0.6))              ##For solving Overfitting problem
  model1.add(Dense(6, activation = "softmax"))
  
  return model1
```

```python
model1 = build_model()
model1.compile(optimizer=Adam(learning_rate=0.001),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
History = model1.fit(XX_train,yy_train, batch_size=100,
                      epochs = 30, validation_data = (X_test,yy_test),
                      verbose = 1 )
```

```python
import matplotlib.pyplot as plt 
plt.plot(History.history['loss'])
plt.plot(History.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend(['train', 'test'])
plt.show()
```

```python
plt.plot(History.history['acc'])
plt.plot(History.history['val_acc'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['train', 'test'])
plt.show()
```

VGG16

```python
from keras.models import Model
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
```

```python
# add preprocessing layer to the front of VGG
vgg = VGG16(input_shape=[224,224,3], weights='imagenet', include_top=False)

# don't train existing weights
for layer in vgg.layers:
  layer.trainable = False
```

```python
# our layers - you can add more if you want
x = Flatten()(vgg.output)
x = Dropout(0.6)(x)
prediction = Dense(6, activation='softmax')(x)

# create a model object
model2 = Model(inputs=vgg.input, outputs=prediction)
```

```python
# view the structure of the model
model2.summary()
```

```python
# tell the model what cost and optimization method to use
model2.compile(
  loss='categorical_crossentropy',
  optimizer=Adam(learning_rate=0.001),
  metrics=['accuracy']
)
```

```python
XX_train = XX_train/255.0
X_test = X_test/255.0
```

```python
r = model2.fit(XX_train,yy_train, batch_size=100,
                      epochs = 5, validation_data = (X_test,yy_test),
                      verbose = 1)
```

```python
import matplotlib.pyplot as plt 
plt.plot(r.history['loss'])
plt.plot(r.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend(['train', 'test'])
plt.show()
```
